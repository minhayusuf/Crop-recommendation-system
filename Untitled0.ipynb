# Nitrogen Analysis

```python
crop_summary_N = crop_summary.sort_values(by='N', ascending=False)

fig = make_subplots(rows=1, cols=2)

top = {
    'y' : crop_summary_N['N'][0:10].sort_values().index,
    'x' : crop_summary_N['N'][0:10].sort_values()
}

last = {
    'y' : crop_summary_N['N'][-10:].index,
    'x' : crop_summary_N['N'][-10:]
}
fig.add_trace(
    go.Bar(top,
           name="Most nitrogen required",
           marker_color=random.choice(colorarr),
           orientation='h',
          text=top['x']),

    row=1, col=1
)

fig.add_trace(
    go.Bar(last,
           name="Least nitrogen required",
           marker_color=random.choice(colorarr),
           orientation='h',
          text=last['x']),
    row=1, col=2
)
fig.update_traces(texttemplate='%{text}', textposition='inside')
fig.update_layout(title_text="Nitrogen (N)",
                  plot_bgcolor='white',
                  font_size=12,
                  font_color='black',
                 height=500)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

# **Splitting dataset into testing and training**

```python
x=crop.drop('label',axis=1)
y=crop['label']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
l=[y_test,y_train]
for i in l:
  print(i.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
x_train,y_train=sm.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

# **Standerdization**

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
x_train=sd.fit_transform(x_train)
x_test=sd.fit_transform(x_test)
```

# **Training models**

# **KNeighbors classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,classification_report
```

```python
params = {'n_neighbors': [3, 5, 7, 9, 11], 'weights': ['uniform', 'distance']}
clf = GridSearchCV(KNeighborsClassifier(), params, cv=10, scoring='accuracy')
clf.fit(x_train, y_train)
print(clf.best_params_)
```

```python

accuracy_list=[]
model=[]
```

```python
cls=KNeighborsClassifier(n_neighbors=3,weights='uniform')
cls.fit(x_train,y_train)
predicted_values=cls.predict(x_test)
print(predicted_values)
m=accuracy_score(y_test,predicted_values)
accuracy_list.append(m)
model.append('Kneighborsclassifier')
print("Kneighborsclassifier's accuracy is: ",m*100)
print(classification_report(y_test,predicted_values))
```

# **Random forest classifier**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint as sp_randint
```

```python
param_dist={'max_depth':[3,4,5,6,None],'max_features':sp_randint(1,11),'min_samples_split':sp_randint(2,11),'criterion':['gini','entropy']}
randomcv=RandomizedSearchCV(RandomForestClassifier(),param_distributions=param_dist,cv=3)
randomcv.fit(x_train,y_train)
print(randomcv.best_params_)
```

```python
rfc=RandomForestClassifier(criterion='entropy',max_depth=6,max_features=3,min_samples_split=7)
rfc.fit(x_train,y_train)
predicted_values=rfc.predict(x_test)
print(predicted_values)
m=accuracy_score(y_test,predicted_values)
accuracy_list.append(m)
model.append('Randomforestclassifier')
print("Randomforestclassifier's accuracy is: ",m*100)
print(classification_report(y_test,predicted_values))
```

# **Decision Tree**

```python
from sklearn import tree
```

```python
params={'max_depth':[4,6,8,None],'criterion':['gini','entropy'],'min_samples_split':sp_randint(1,8),'min_samples_leaf':sp_randint(1,5)}
randomcv=RandomizedSearchCV(RandomForestClassifier(),params,cv=10)
randomcv.fit(x_train,y_train)
print(randomcv.best_params_)
```

```python
dt=tree.DecisionTreeClassifier(criterion='entropy',max_depth=None,min_samples_leaf=2, min_samples_split=4)
dt.fit(x_train,y_train)
predicted_values=dt.predict(x_test)
print(predicted_values)
m=accuracy_score(y_test,predicted_values)
accuracy_list.append(m)
model.append('DecisionTreeClassifier')
print("DecisionTreeClassifier's accuracy is: ",m*100)
print(classification_report(y_test,predicted_values))
```

# **Naive bayes**

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
predicted_values=nb.predict(x_test)
print(predicted_values)
m=accuracy_score(y_test,predicted_values)
accuracy_list.append(m)
model.append('Navebayes')
print("Navebayes's accuracy is: ",m*100)
print(classification_report(y_test,predicted_values))
```

```python
plt.figure(figsize=(10,5),dpi=100)
plt.grid(False)
plt.title('Accuracy Comparison')
plt.xlabel('accuracy')
plt.ylabel('model')
sns.barplot(x=accuracy_list,y=model,palette=['blue','green','red','black'])
for i, accuracy in enumerate(accuracy_list):
    plt.text(accuracy, i, f'{accuracy:.2f}%', ha='left', va='center', fontsize=12, color='black')
plt.show()
```

```python
accuracy_models=dict(zip(model,accuracy_list))
for i,j in accuracy_models.items():
  print(i,':',j)
```
